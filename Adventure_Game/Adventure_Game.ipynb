```python
name = input("Type Your Name: ")
print("Welcome ", name, "to this Adventure")

answer = input("You spawn in the middle city. You can go right, left or forward. Which way would you like to go? (Left/Right/Forward)").lower()

if answer == "left":
  answer = input("You come to the mountain. In the mountain you saw a Goblin. DO you want to attack or run away? (Attack/Run)").lower()

  if answer == "attack":
    print("You got slash by the goblin. You Lose")
  elif answer == "run":
    answer = input("You go back to the city. Now which way would you like to go? (Right/Forward)").lower()
  else:
    print("Not Valid. You Lose")

elif answer == "right":
  answer = input("You come to the river. Do you want to accross or just go back to the City? (Accross/Back)").lower()

  if answer == "accross":
    print("You swam accross and got eaten by the alligator. You Lose")
  elif answer == "Back":
    answer = input("You go back to the city. Now which way would you like to go? (Forward)").lower()
  else:
    print("Not Valid. You Lose")

elif answer == "forward":
  answer = input("You come to the forest. In the forest you can walk to the right or to the left. Which way would you like to go? (Right/Left)").lower()

  if answer == "right":
    print("You meet a Dragon. Dragon use fireballs and kill you. You Lose")
  elif answer == "left":
    print("You got a treasure in the ruin. You Win")
  else:
    print("Not Valid. You Lose")

else:
  print("Not Valid. You Lose")
```
