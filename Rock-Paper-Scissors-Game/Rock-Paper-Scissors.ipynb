```python
import random

print("ROCK-PAPER-SCISSORS GAME")
print("--------------------------------"+"\n")

exit = False
user_score = 0
computer_score = 0

while exit == False:
  options = ["Rock", "Paper", "Scissors"]
  user_input = input("Choose Rock, Paper, Scissors or exit: ")
  computer_input = random.choice(options)
  valid = "rock" or"Rock" or "Paper" or"paper" or "Scissors" or "scissors" or "Exit" or "exit"

  if user_input == "Exit" or user_input == "exit":
    print("Game Ended")
    print("Your Score: " + str(user_score))
    print("Computer Score: "+ str(computer_score))
    exit = True
    if user_score < computer_score:
      print(">>>>> You Lose the Game <<<<<")
    elif user_score > computer_score:
      print(">>>>> You Win the Game <<<<<")
    elif user_score == computer_score:
      print(">>>>> It is Draw <<<<<Pape")

  if user_input == "Rock" or user_input == "rock":
    if computer_input == "Rock":
      print("You choose Rock")
      print("Computer choose Rock")
      print("It is a tie" + "\n")

    elif computer_input == "Paper":
      print("You choose Rock")
      print("Computer choose Paper")
      print("Computer Wins" + "\n")
      computer_score += 1

    elif computer_input == "Scissors":
      print("You choose Rock")
      print("Computer choose Scissors")
      print("You Wins" + "\n")
      user_score += 1

  if user_input == "Paper" or user_input == "paper":
    if computer_input == "Rock":
      print("You choose Paper")
      print("Computer choose Rock")
      print("You Win" + "\n")
      user_score += 1

    elif computer_input == "Paper":
      print("You choose Paper")
      print("Computer choose Paper")
      print("It is a tie" + "\n")

    elif computer_input == "Scissors":
      print("You choose Paper")
      print("Computer choose Scissors")
      print("Computer Win" + "\n")
      computer_score += 1

  if user_input == "Scissors" or user_input == "scissors":
    if computer_input == "Rock":
      print("You choose Scissors")
      print("Computer choose Rock")
      print("Computer Win" + "\n")
      computer_score += 1

    elif computer_input == "Paper":
      print("You choose Scissors")
      print("Computer choose Paper")
      print("You Win" + "\n")
      user_score += 1

    elif computer_input == "Scissors":
      print("You choose Scissors")
      print("Computer choose Scissors")
      print("It is a tie" + "\n")

  elif user_input.lower() not in ["rock", "paper", "scissors", "exit"]:
    print("Invalid Input")
```
